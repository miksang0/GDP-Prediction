```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#reading data
data = pd.read_csv('/content/imf-dm-export-20240114.csv')
```

```python
#view data
data.info()
```

```python
sns.regplot(x='YEAR', y = 'Nepal', data=data)
#data visualization
```

```python
 #for regression
 #create an instance
lri = LinearRegression()
```

```python
#train model
lri.fit(data[['YEAR']].values, data.Nepal)
```
